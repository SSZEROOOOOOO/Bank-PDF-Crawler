```python
import os
import glob
import tabula
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
pdf_path = "../pdfs"
pdfs = glob.glob("{}/*.pdf".format(pdf_path))
```

```python
dfk = pd.read_csv('../keywords.csv', header=None, sep='\t')
keywords = dfk[0].values.tolist()
```

```python
dff=pd.DataFrame({0 : ["产品名称","理财信息登记系统登记编码","产品类型","初始计划发行量","业绩比较基准","托管费率",
                            "销售手续费率","固定管理费率","认购费率","浮动管理费率" ,"投资策略","销售范围","投资管理人"]})
```

```python
mergeDF = dff
```

```python
for pdf in pdfs:
    dfs = tabula.read_pdf(pdf, encoding='utf-8', 
                          lattice=True, stream=True, guess=False, pages='all', multiple_tables=True, pandas_options={'header':None})
    df0 = pd.DataFrame()
    df0 = df0.sort_index()
    print('Extracting',pdf)
    for i in range(len(dfs)):
        dfs[i][0]=dfs[i][0].astype(str)
        df0 = df0.append(dfs[i][dfs[i][0].str.contains('|'.join(keywords), na=False)], ignore_index=True)
        df0.dropna(axis="columns", how="all", inplace=True)
        #df0[0]= df0[0].str.split('费').str[0]
        df0=df0[~df0[0].str.contains("理财非存款")]
        df0[0]=df0[0].apply(lambda x : [process.extract(x,dff[0], limit=1)][0][0][0])
        df0.drop_duplicates(0, keep='first', inplace=True)
        df0.dropna(axis="columns", how="all", inplace=True)
    mergeDF = mergeDF.merge(df0, how='left', on = 0)
    print()
```
